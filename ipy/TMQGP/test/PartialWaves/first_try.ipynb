```python
import TMQGP as tm
import QuarkTM
from numpy import *
```

```python
qrange = linspace(0, 5, 51)
erange = linspace(-5, 5, 201)
```

```python
p = QuarkTM.Particle(0.5, qrange, erange)
```

```python
ch = QuarkTM.Channel(p, p, 0.2)
```

```python
ch.populate_T()
```

```python
iEps = tm.Interpolator(qrange, p.om0(qrange), 'linear')
```

```python
S_l0 = array([tm.sigma_ff_onshell(e, .25, 0.2, ch.iImT, p.R, iEps, iEps) for e in erange])
```

```python
S_l1 = array([tm.sigma_ff_onshell(e, .25, 0.2, ch.iImT, p.R, iEps, iEps, 0) for e in erange])
```

```python
plt.plot(erange, S_l0)
plt.plot(erange, S_l1)
# plt.plot(erange, S_l0 + S_l1)
```
