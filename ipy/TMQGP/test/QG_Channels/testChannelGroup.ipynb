```python
from numpy import *
```

```python
import QuarkTM
```

```python
params = [
    {'G' : 8, 'L': 0.5, 'screen' : 0.05},
    {'G' : 8, 'L': 0.5, 'screen' : 0.05},
]
```

```python
qrange = linspace(0, 5, 51)
erange = linspace(-5, 5, 251)
```

```python
q = QuarkTM.Particle(0.5, qrange, erange)
```

```python
chl = QuarkTM.ChannelL('qa1', 1, q, q, 0.2, params, ds=4, da=1, Fa=1)
```

```python
chl.populate_T()
```

```python
ch = QuarkTM.Channel(q, q, 0.2, **params[0], ds=4, da=1, Fa=1)
```

```python
ch.populate_T()
```

```python
plt.plot(erange, imag(ch.TM[:, 0]))
plt.plot(erange, imag(chl.chs[0].TM[:, 0]))
```

```python
plt.plot(erange, imag(chl.chs[0].TM[:, 1]))
plt.plot(erange, 3*imag(chl.chs[1].TM[:, 1]))
```

```python
TM = chl.get_TM()
```

```python
plt.plot(erange, imag(TM[:, 0]))
```

```python
cg = QuarkTM.ChannelGroup()
```

```python
cg.addChannel(chl)
```

```python
TM = cg.get_T()
```

```python
plt.plot(erange, imag(TM[:, 0]))
```
