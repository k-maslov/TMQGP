```python
erange = linspace(-5, 5, 201)
qrange = linspace(0, 5, 151)
```

```python
q = QuarkTM.Particle(0.5, qrange, erange, eps=2e-2)
```

```python
ch = QuarkTM.Channel(q, q, 0.15, G=6)
```

```python
xs = np.array([[tm.x_solve(E, k, k, ch.T, ch.iV, ch.iOm, ch.iReG2, ch.iImG2, 5,
            int(np.sign(ch.G))) for k in ch.qrange]
               for E in ch.erange])
```

```python
ch.populate_T()
```

```python
plt.plot(erange, ch.ImG2[:, 0])
```

```python
plt.plot(erange, imag(xs[:, 0]))
```

```python
plt.plot(erange, imag(l[:, 0]))
```

```python
LT = v1v2 * -l
```
