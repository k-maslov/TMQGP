```python
import TMQGP as tm
import QuarkTM
from numpy import *
```

```python
qrange = linspace(0, 5, 51)
erange = linspace(-5, 5, 251)
```

```python
pt = QuarkTM.Particle(0.5, qrange, erange, eps=.5e-2)
```

```python
ch1 = QuarkTM.Channel(pt, pt, 0.2, G=6)
ch2 = QuarkTM.Channel(pt, pt, 0.2, G=-6)
```

```python
plt.plot(qrange, [ch2.iV(q) for q in qrange])
```

```python
ch1.populate_T()
ch2.populate_T()
```

```python
tm.T_solve(0.3, 0, 0, 0.2, ch1.iV, ch1.iOm, ch1.iReG2, ch1.iImG2, )
```

```python
plt.plot(erange, imag(ch1.TM[:, 0]))
plt.plot(erange, imag(ch2.TM[:, 0]), ls=':')
```
