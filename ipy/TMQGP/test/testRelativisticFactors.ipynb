```python
import TMQGP as tm
import QuarkTM
from numpy import *

qrange = linspace(0, 5, 51)
erange = linspace(-5, 5, 251)
```

```python
pt = QuarkTM.Particle(0.5, qrange, erange, eps=1e-2)
```

```python
ch1 = QuarkTM.Channel(pt, pt, 0.2, do_rel=1)
ch2 = QuarkTM.Channel(pt, pt, 0.2, do_rel=0)
```

```python
ch1.populate_T()
ch2.populate_T()
```

```python
plt.plot(erange, imag(ch1.TM[:, 0]))
plt.plot(erange, imag(ch2.TM[:, 0]), ls=':')
```
