```python
import TMQGP as tm
import QuarkTM
from numpy import *
```

```python
import matplotlib 
matplotlib.style.use('publication23')
```

```python
qrange = linspace(0, 5, 51)
erange = linspace(-5, 5, 251)
```

#  QQ

```python
pt = QuarkTM.Particle(0.5, qrange, erange, eps=2e-2)
```

```python
ch1 = QuarkTM.Channel(pt, pt, 0.2, G=6)
ch2 = QuarkTM.Channel(pt, pt, 0.2, G=-6)
```

```python
ch1.populate_T()
ch2.populate_T()
```

```python
ch1.populate_S()
```

```python
plt.plot(ch1.erange, imag(ch1.TM[:, 0]))
```

```python
plt.plot(erange, ch1.ImS[:,0])
```
