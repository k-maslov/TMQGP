# GG

```python
g = QuarkTM.Particle(1., qrange, erange, eps=2e-2, stat='b', d=16,
                    propagator=1)
```

```python
ch_g = QuarkTM.Channel(g, g, 0.2, G=8, L=0.5, screen=1.5, calc=2, 
                       Fa=9/4, ds=9)
```

```python
ch_g.populate_T()
```

```python
ch_g.populate_S()
```

```python
plt.plot(erange, imag(ch_g.TM[:, 0]))
```

```python
plt.plot(erange, ch_g.ImS[:, 0])
# plt.ylim(-0.01, 1e-5)
```

## Plotting the integrand

```python
p = 0.01
om = -.5
iImT = ch_g.iImT
T = 0.2
```

```python
ints = array([
    tm.k_integral_cm(e, om, p, ch_g.iImT, g.iImG) * 
    (tm.n_b(e - om, T) - tm.n_b(e, T))
    for e in erange])
```

```python
plt.plot(erange, ints)
```
