## Semi-analytic integral

```python
from scipy.integrate import quad
```

```python
E = 0.5
```

```python
def re_an(E):
    if E > 2*mQ:
        return quad(lambda z: 2/pi * z**2 * iV(z)**2, 0, 5, weight='cauchy', wvar=sqrt(E**2 - mQ**2))[0]
    else:
        return quad(lambda z: real(2/pi * z**2 * iV(z)**2 / (2*sqrt(mQ**2 + z**2) - E + 2j*0.001)), 0, 5)[0]
```

```python
def re_an_G(E):
    if E > 2*mG:
        return quad(lambda z: 2/pi * z**2 * iV(z)**2, 0, 5, weight='cauchy', wvar=sqrt(E**2 - mG**2))[0]
    else:
        return quad(lambda z: real(2/pi * z**2 * iV(z)**2 / (2*sqrt(mG**2 + z**2) - E + 2j*0.001)), 0, 5)[0]
```

```python
res_an = array([re_an(E) for E in erange_T])
```
