$$\rho(\omega, q) = A e^{-(\omega - \omega_q)^2 / 2 \Gamma^2}$$

$$\int\limits_{-\infty}^{+\infty} \rho(\omega, q) d\omega = 1 = A \int\limits_{-\infty}^{+\infty} d\omega e^{-(\omega - \omega_q)^2 / 2 \Gamma^2} = A \sqrt{\pi / (1 / 2 \Gamma^2)}$$ 

$$A = \sqrt{\frac{1}{2 \pi \Gamma^2}}$$

```python
from scipy.integrate import quad
```

```python
m = 0.5
```

```python
from numpy import *
```

```python
def rho(om, q, gamma=0.1):
    return sqrt(1/2  / pi / gamma**2) * exp(-(om - sqrt(q**2 + m**2))**2 / 2 / gamma**2)
```

```python
omrange = linspace(-5, 5, 500)
```

```python
plt.plot(omrange, rho(omrange, 0))
```

```python
quad(lambda z: rho(z, 0), -5, 5)
```
